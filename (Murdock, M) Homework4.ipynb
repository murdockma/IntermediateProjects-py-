```python
# here are the point values for each letter
letterValues = {'a':1,'b':3,'c':3,'d':2,'e':1,'f':4,'g':2,'h':4,'i':1,'j':8,'k':5,'l':1,'m':3,
                 'n':1,'o':1,'p':3,'q':10,'r':1,'s':1,'t':1,'u':1,'v':4,'w':4,'x':8,'y':4,'z':10}
```

```python
# which word that starts with the letter Z has the most points?

# \\b starts word boundary followed by Z (the first character we want to match), 
## followed by \w+ (matches 1 or more alphanumerics) to get the rest of the letters in the word
all_z = re.findall('\\bZ\w+', sowpods)
cit = {} #empty dict

# iterate through our findall list, assign the word as the key and the sum of the word as the value of the key
for word in all_z:
    # retrieves value associated with character(iteration) from the letterValues dict and sums the characters values
    cit[word] = sum(letterValues[char] for char in word.lower()) 

# iterate through key,value pair in the dicts items and check if value(iteration) is equal to the max value 
## in the dictionaries values. If so, print the key associated with that val and the val itself
for key, val in cit.items():
    if val == max(cit.values()):
        print(f'The highest scoring word that stars with the letter Z is {key}. It is worth {val} points.')
```

```python
# how many words start with q and end with a vowel?
# use a regular expression to find the answer

# len gets count of pattern. \\b specifies beginning of word boundary, followed by [Qq] which matches either Q or q
## the \S+ matches 1 or more non-whitespace characters (\w+ would work here as well). 
### Then the [aeio...] matches any one vowel from the list followed by the end of the word boundary \\b (ends pattern)
len(re.findall('\\b[qQ]\S+[aeiouAEIOU]\\b', sowpods))
```

```python
# how many words have a q not followed by u?
# use a regular expression to find the answer


# \w* matches 0 or more alphanumeric characters before the next specified character (Qq)
## [Qq] matches either Q or q, followed by a negative lookahead, which matches 
### anything other than the character U. The \w* at the end matches 0 or more alphanumerics after the (?!U)
### the len() gets count of findall list
qNotU = len(re.findall('\w*[Qq](?!U)', sowpods))

#print count
print(f'There are {qNotU} words that have a Q not followed by a U.')
```

```python
# what are the words that have a q not followed by a u?

# \w* matches 0 or more alphanumerics before the next specified character is matched (Qq)
## [Qq] matches either Q or q, followed by a negative lookahead, which matches 
### anything other than the character U. The \w* at the end matches 0 or more alphanumerics after the (?!U)
re.findall('\w*[Qq](?!U)\w*', sowpods)
```

```python
# how many words start with a vowel and end with a vowel?
# use a regular expression to find the answer

# \\b specifies beginning of word boundary, followed by [aeio...] which matches any of the characters in the list(vowels).
## The \w* after will match any 0 or more alphanumerics between the first vowel and second character 
###specified(another list of vowels), followed by the end of word boundary \\b
vowCnt = len(re.findall("\\b[aeiouAEIOU]\w*[aeiouAEIOU]\\b",sowpods))

print(f'There are {vowCnt} words that start and end with a vowel.')
```

```python
# what's the longest word that starts with a vowel and ends with a vowel?

# \\b specifies beginning of word boundary, followed by [aeio...] which matches any of the characters in the list(vowels).
## The \w* after will match any characters between the first vowel and second character specified(which is another vowel),
### followed by the end of word boundary \\b
start_end = re.findall('\\b[aeiouAEIOU]\w*[aeiouAEIOU]\\b', sowpods)

#empty string
st=''
# iterate through findall list and add each iteration to a empty string with a space after
for word in start_end:
    st+=word+' '
    
# get longest word based on its length (.split())(key=len)
mxx = max(st.split(), key=len)

# print the longest word, and its length
print(f'The longest word that starts with a vowel and ends with a vowel is {mxx}. It has {len(mxx)} letters.')
```

```python
# what words start with an X and end with an X?

# \\b specifies beginning of word boundary, followed by X, which matches any character 'X', followed by \w*
## which matches 0 or more alphanumeric letters between X and the next letter specified(X). 
## The \\b at the end of the pattern specifies end of word boundary (after character X)
re.findall('\\bX\w*X\\b', sowpods)
```

```python
# open file and store all text in a variable
with open('AllBeatlesLyrics.txt') as beatle:
    beatleText = beatle.read()
```

```python
# if you look at the file contents, you'll notice several \xa0 characters which is a non-breaking space.  
# demonstrate how you would change these non-breaking spaces to normal spaces.
# note: if you don't see these non-breaking space characters, that's okay, but you'll still need to write the
# code to accomplish this task.

# replace all insatnces of '\xa0' with ' '(space)
beatleText=beatleText.replace("\xa0", ' ') 
```

```python
# how many times is the word "love" or "Love" used in the lyrics?
# use a regular expression

# [Ll]ove matches all words that contain Love or love. len() gets count of findall list
love_cnt = len(re.findall('[Ll]ove', beatleText))

# print count
print(f'The word love is used {love_cnt} times in these lyrics.')
```

```python
# using regular expressions, how many phrases are there that start with the word "love"
# or "Love"? Consider a phrase to be two or more words that ends with any punctuation mark 
# but does not include the punctuation.

# what is the longest phrase containing the word love and how many words are in it?


# [Ll] matches either characters L or l, followed by 'ove' to match either Love or love.
## The \w* after love mathches 0 or more alphanumerics (ex. loves, lovers, etc).
### [\w\s]+ matches whitespace and alphanumeric charcters and continues the pattern until it reaches punctuation
allPhrases = re.findall("[Ll]ove\w* [\w\s]+", beatleText)

# printing length of findall list
print(f'There are {len(allPhrases)} words that start with the word love in these lyrics.')

new_d = {} # empty Dict

# iterate through findall list and assign phrases as keys, and the length 
## of the split phrases(so we can count words, not characters) as values
for phrase in allPhrases:
    new_d[phrase] = len(phrase.split())

# iterate throuh key,val of dictionary (in item pair) and check if the val(length of phrases) 
## is equal to the max val in the dictionaries values. If it is, print the key associated with the value,
### and print the value itself
for key,val in new_d.items():
    if val == max(new_d.values()):
        print(f'The longest phrase in the word love is:\n{key}\nIt contains {val} words.')
```

```python
# What are the phrases (use the same definition as above) that start with the word beep?
# use a regular expression

## [Bb] matches either characters B or b, followed by 'eep' to match eihter Beep or beep. 
## The \w* matches 0 or more alphanumerics atfter the specified word in the pattern (ex. beep, beeps..)
### [\w\s]+ matches whitespace and alpha numeric charcters and continues the pattern until it reaches punctuation
beepPhrases = re.findall("[Bb]eep\w* [\w\s]+", beatleText)

#print findall list
print(f'The phrases that start with the word beep are\n{beepPhrases}')
```
